```python
import matplotlib.pyplot as plt


color_list = [[0, 0, 0], [238,172,255], [240,130,40], [251,111,111], [121, 173, 214]]

def add_color(mask):
    color_mask = np.zeros((mask.shape[0], mask.shape[1], 3))   
    for i in range(5):
        color_mask[mask == i, :] = np.array(color_list[i]) / 255

    color_mask[mask == 255, :] = np.array([255, 255, 255]) / 255
    return color_mask

plt.figure(figsize=(25, 20))
clsout[clsout == 255] = 0
clsgloout[clsgloout == 255] = 0
cnt = 1
for i in range(16):
    plt.subplot(8, 12, cnt)
    plt.imshow(add_color(gt[i]), vmin=0, vmax= 6)
    plt.axis('off')
    cnt += 1
    plt.subplot(8, 12, cnt)
    plt.imshow(entro[i])
    plt.axis('off')
    cnt += 1
    plt.subplot(8, 12, cnt)
    plt.imshow(add_color(oriout[i]), vmin=0, vmax= 6)
    plt.axis('off')
    cnt += 1
    plt.subplot(8, 12, cnt)
    plt.imshow(add_color(clsout[i]), vmin=0, vmax= 6)
    plt.axis('off')
    cnt += 1
    plt.subplot(8, 12, cnt)
    plt.imshow(add_color(clsgloout[i]), vmin=0, vmax= 6)
    plt.axis('off')
    cnt += 1
    plt.subplot(8, 12, cnt)
    plt.imshow(add_color(clsout[i]-clsgloout[i]), vmin=0, vmax = 1)
    plt.axis('off')
    # print(np.unique(clsgloout[i]-clsout[i]), (clsgloout[i]-clsout[i]).sum())
    cnt += 1


# plt.savefig('test.png')
# plt.close()
```
